```python
# Definir los rangos etarios:
bins = [0, 20, 30, 40, 50, 60, 70, 80]
labels = ['0-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79']

# Agrupar las edades en los rangos definidos:
df_SelectFinal['AgeRange'] = pd.cut(df_SelectFinal['Age'], bins=bins, labels=labels, right=False)

#Contar cuántas personas hay en cada rango:
age_counts = df_SelectFinal['AgeRange'].value_counts(sort=False)

#Calcular los porcentajes:
age_counts_percentage = 100 * age_counts / age_counts.sum()

plt.figure(figsize=(10, 10))
plt.pie(age_counts_percentage, labels=age_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución porcentual de rangos Etarios', y=1.1)
plt.axis('equal')  # Para asegurar que el gráfico sea un círculo
plt.show()
```

### Se procede a realizar los graficos las presiones del dia

### Grafico de dispersión  (Sistolica y Diastolica) por cada uno de los pacientes

```python
# Seleccionar datos de un paciente específico
data = patientStudies[patient]

# Convertir los valores de tiempo a horas del día
timeHours = [(t.hour + t.minute / 60 + t.second / 3600) for t in data['Time']]

# Ordenar los datos por la hora en horas del día
dataSorted = sorted(zip(timeHours, data['Systolic'], data['Diastolic']))

# Desempaquetar los datos ordenados
timeSorted, systolicSorted, diastolicSorted = zip(*dataSorted)

# Variables de control para las etiquetas
systolicA120 = False
systolicB120 = False
diastolicA80 = False
diastolicB80 = False

# Configurar el tamaño de la figura
plt.figure(figsize=(12, 6))

# Si la presión sistólica supera 120, graficar en rojo, de lo contrario, en azul
for i in range(len(timeSorted)):
    if systolicSorted[i] > 120:
        if not systolicA120:
            plt.plot(timeSorted[i], systolicSorted[i], 'ro', label='Sistólica > 120')
            systolicA120 = True
        else:
            plt.plot(timeSorted[i], systolicSorted[i], 'ro')
    else:
        if not systolicB120:
            plt.plot(timeSorted[i], systolicSorted[i], 'bo', label='Sistólica ≤ 120')
            systolicB120 = True
        else:
            plt.plot(timeSorted[i], systolicSorted[i], 'bo')

# Si la presión diastólica supera 80, graficar en amarillo, de lo contrario, en verde        
for i in range(len(timeSorted)):
    if diastolicSorted[i] > 80:
        if not diastolicA80:
            plt.plot(timeSorted[i], diastolicSorted[i], 'yo', label='Diastólica > 80')
            diastolicA80 = True
        else:
            plt.plot(timeSorted[i], diastolicSorted[i], 'yo')
    else:
        if not diastolicB80:
            plt.plot(timeSorted[i], diastolicSorted[i], 'go', label='Diastólica ≤ 80')
            diastolicB80 = True
        else:
            plt.plot(timeSorted[i], diastolicSorted[i], 'go')        


plt.xlabel('Tiempo (Horas del Día)')
plt.ylabel('Presión')
plt.title('Gráfico de Presión Arterial')
plt.legend()
plt.grid(True)
plt.show()
```

### Grafico Lineal (Sistolica y Diastolica)

```python
# Crear el gráfico
plt.figure(figsize=(12, 6))

# Graficar la presión diastólica
plt.plot(timeSorted, diastolicSorted, label='Diastólica', color='green')

# Graficar la presión sistólica con coloración condicional
for i in range(1, len(timeSorted)):
    plt.plot(timeSorted[i-1:i+1], systolicSorted[i-1:i+1], color='red' if systolicSorted[i] > 120 else 'blue')

# Configurar título y etiquetas
plt.title('Gráfico de Presión Arterial')
plt.xlabel('Tiempo (Horas del Día)')
plt.ylabel('Presión')
plt.legend(['Diastólica', 'Sistólica ≤ 120', 'Sistólica > 120'])
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

### Grafico de Barra (Sistolica y Diastolica)
#### Se muestra el Promedio de la disminucion de algunas de las presiones entre el dìa y la noche de todos los pacientes

```python
# Eliminar registros donde Systolic y Diastolic sean 0
dfr = pd.read_csv(ruta+"TestsRawData-DatosLimpios.csv" )
dfr = dfr[(dfr['Systolic'] != 0) & (dfr['Diastolic'] != 0)]

# Eliminar las columnas 'Date' y 'Time'

dfr = dfr.drop(columns=['Date', 'Time'])

# Mostrar el DataFrame resultante
print(dfr)
```

```python
#Grafico que muestre El Promedio de la disminucion de algunas de las presiones entre el dìa y la noche de todos los pacientes

# Filtrar los datos para eliminar registros donde Systolic y Diastolic sean 0
dfr = dfr[(dfr['Systolic'] != 0) & (dfr['Diastolic'] != 0)]


# Convertir la columna 'ReadingTime' a tipo datetime con el formato correcto
dfr['ReadingTime'] = pd.to_datetime(dfr['ReadingTime'], format='%d/%m/%Y %H:%M:%S')

# Función para determinar si es de día o de noche
def es_de_dia(hora):
    return hora.hour >= 6 and hora.hour < 18

# Aplicar la función y crear una nueva columna 'DayNight'
dfr['DayNight'] = dfr['ReadingTime'].apply(es_de_dia)

# Filtrar los datos para obtener solo las lecturas durante el día y la noche
lecturas_dia = dfr[dfr['DayNight'] == True]
lecturas_noche = dfr[dfr['DayNight'] == False]

# Calcular la diferencia entre las lecturas de presión sistólica y diastólica entre el día y la noche
dif_sistolica = lecturas_dia['Systolic'].mean() - lecturas_noche['Systolic'].mean()
dif_diastolica = lecturas_dia['Diastolic'].mean() - lecturas_noche['Diastolic'].mean()

# Crear un gráfico de barras apiladas
plt.figure(figsize=(12, 6))
categorias = ['Systolic', 'Diastolic']
diferencias_dia = [lecturas_dia['Systolic'].mean(), lecturas_dia['Diastolic'].mean()]
diferencias_noche = [lecturas_noche['Systolic'].mean(), lecturas_noche['Diastolic'].mean()]

plt.bar(categorias, diferencias_dia, label='Día', color= 'skyblue')
plt.bar(categorias, diferencias_noche, bottom=diferencias_dia, label='Noche', color= 'lightgreen')
plt.xlabel('Presión')
plt.ylabel('Promedio de presión')
plt.title('Promedio de presión durante el día y la noche')
plt.legend()
plt.show()
```

### Grafico de Cajas  (Sistolica y Diastolica)
#### Se visualiza las presiones promedias de dia y de noche.

```python
#Grafico 2 que muestre El Promedio de la disminucion de algunas de las presiones entre el dìa y la noche de todos los pacientes.

# Filtrar los datos para eliminar registros donde Systolic y Diastolic sean 0
dfr = dfr[(dfr['Systolic'] != 0) & (dfr['Diastolic'] != 0)]

# Convertir la columna 'ReadingTime' a tipo datetime con el formato correcto
dfr['ReadingTime'] = pd.to_datetime(dfr['ReadingTime'], format='%d/%m/%Y %H:%M:%S')

# Definir la función para determinar si es de día o de noche
def es_de_dia(hora):
    return 6 <= hora.hour < 18

# Aplicar la función y crear una nueva columna 'DayNight'
dfr['DayNight'] = dfr['ReadingTime'].apply(es_de_dia)

# Crear el gráfico de caja
plt.figure(figsize=(14, 7))

# Datos para el gráfico de caja
data = [dfr[dfr['DayNight'] == True]['Systolic'], dfr[dfr['DayNight'] == False]['Systolic'],
        dfr[dfr['DayNight'] == True]['Diastolic'], dfr[dfr['DayNight'] == False]['Diastolic']]

# Etiquetas para el gráfico
labels = ['Sistólica Día', 'Sistólica Noche', 'Diastólica Día', 'Diastólica Noche']

# Crear el box plot sin mostrar los valores atípicos
box = plt.boxplot(data, labels=labels, patch_artist=True, showfliers=False)

# Color de las cajas
colors = ['skyblue', 'lightgreen', 'lightcoral', 'lightyellow']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

# Personalización del gráfico
plt.xlabel('Promedio de Presión', fontsize=14)
plt.ylabel('Presión (mm Hg)', fontsize=14)
plt.title('Distribución de Presión Sistólica y Diastólica durante el Día y la Noche', fontsize=16)

# Mejorar la rejilla
plt.grid(True, linestyle='--', linewidth=0.5, color='blue')

# Mostrar el gráfico
plt.show()
```

El gráfico de caja te permite ver las diferencias de las lecturas de presión arterial que varían entre el día y la noche para todos los pacientes. La caja está delimitada por el primer cuartil (Q1) en la parte inferior y el tercer cuartil (Q3) en la parte superior. La línea dentro de la caja representa la mediana (Q2) del conjunto de datos. Los bigotes se extienden desde los cuartiles hasta el valor más alto y bajo dentro de 1.5 veces el rango intercuartil (IQR). Si la línea central de la caja de "Sistólica Día" está más alta que la de "Sistólica Noche", significa que la presión sistólica tiende a ser mayor durante el día. Esto sugiere que, en general, las presiones sistólica y diastólica son más altas y varían más durante el día, mientras que durante la noche tienden a ser más bajas y menos variables, aunque con algunos valores atípicos.


### Grafico de Histograma 
#### Distribucion de Frecuencia para Presion Sistólica

```python
#Histograma de Distribucion de Frecuencia para Presion Sistólica
# Filtrar los datos para obtener solo las lecturas durante el día y la noche
lecturas_dia = dfr[dfr['DayNight'] == True]
lecturas_noche = dfr[dfr['DayNight'] == False]

# Crear histogramas comparativos de las presiones sistólicas máximas durante el día y la noche
plt.figure(figsize=(14, 7))

# Histograma para presión sistólica
plt.hist(lecturas_dia['Systolic'], bins=20, alpha=0.5, label='Sistólica Día', color='blue', edgecolor='black')
plt.hist(lecturas_noche['Systolic'], bins=20, alpha=0.5, label='Sistólica Noche', color='green', edgecolor='black')

# Personalización del gráfico
plt.xlabel('Presión Sistólica (mm Hg)', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)
plt.title('Distribución de Presión Sistólica durante el Día y la Noche', fontsize=16)
plt.legend(loc='upper right')
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

Estas Interpretaciónes de Histograma de Distribución de Frecuencia: Se crean dos histogramas superpuestos, uno para las lecturas de presión sistólica durante el día y otro para la noche.
El histograma muestra cuántas veces ocurren ciertas lecturas de presión sistólica durante el día y la noche.
Comparación de Lecturas de Día y Noche: Comparando las barras de los histogramas, puedes observar si las lecturas de presión sistólica tienden a ser más altas o más bajas en ciertos rangos durante el día o la noche.
Transparencia y Superposición: La superposición con diferentes colores y transparencias permite ver claramente las diferencias y similitudes en la distribución de las lecturas.
